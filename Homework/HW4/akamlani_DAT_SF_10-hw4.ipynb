## Dimension Reduction, Clustering Techniques

```python
# Dataset: http://scikit-learn.org/stable/datasets/olivetti_faces.html#olivetti-faces
from pprint import pprint
import numpy as np
from sklearn.datasets import fetch_olivetti_faces 

ol_faces = fetch_olivetti_faces()
#print(ol_faces.DESCR)

# 10 images of 40 different people = 40 x 10 =400
# Each row corresponds to image of original size 64 x 64 pixels
X = ol_faces.data
images = ol_faces.images
# Target Labels ranging from 0-39 and correspond to the Subject IDs
y = ol_faces.target
n_samples, n_features = X.shape
print "Dimensions -> Data:{0}, Images:{1}, Target:{2}".format(X.shape, images.shape, y.shape)
```

## Question 1. Implement PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
X_r = pca.fit(X).transform(X)
# percentage of variance explained by each of selected components
print"Variance Ratio: \n{0}".format(pca.explained_variance_ratio_[:15])
# components with maximum variance
print"Components...: \n{0}".format(pca.components_[:1][0][:10])
```

```python
# identify where the elbow is (scee plot)
import matplotlib.pyplot as plt
%matplotlib inline

def plot_comp_analysis(pca_v):
    comp_id = [i + 1 for i in range(len(pca_v.explained_variance_ratio_))]             
    fig = plt.figure(figsize=(8,5))
    plt.plot(comp_id, pca_v.explained_variance_ratio_, 'ro-', linewidth=2)
    plt.title('Scree Plot')
    plt.xlim(0,25)
    plt.xlabel('Principal Component')
    plt.ylabel('Proportion of Variance')
    
plot_comp_analysis(pca)
```

```python
# contrast against scaling 
# (Gaussian with zero mean and unit variance)
# re-using current PCA

#from sklearn.preprocessing import scale
#Scaled_X = scale(X, axis=0, with_mean=True, with_std=True, copy=True )
#Scaled_X_r = pca.fit(Scaled_X).transform(Scaled_X)

# Contrast against Normalized
Norm_X = (X - X.mean())/X.std()
Norm_X_r = pca.fit(Norm_X).transform(Norm_X)
# percentage of variance explained by each of selected components
print"Variance Ratio: \n{0}".format(pca.explained_variance_ratio_[:15])
print"Components...: \n{0}".format(pca.components_[:1][0][:10])

plot_comp_analysis(pca)
```

#### Principal Components Analysis
* Early on the it shows us that many of the features are not adding a high correlation, so reducing the number of components would be beneficial (at the elbow, this corresponds to 4-5).

#### Scaling
* Unfortunately it does not look like scaling via normalization resulted in a higher variance.

## Question 4. k-means clustering w/k=number of people in dataset

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances

def kmeans_alg(samples, num_clusters=1, sequence_type='random', seed_iterations=10, 
               max_iterations=300, random_gen=1):
    km = KMeans(n_clusters=num_clusters,  init=sequence_type, 
                n_init=seed_iterations , max_iter = max_iterations, random_state=random_gen)
    km.fit(samples)
    return km
        
def start_kmeans(samples, num_clusters=1, sequence_type='random', seed_iterations=10, 
                 max_iterations=300, random_gen=1):
    km = kmeans_alg(samples, num_clusters, sequence_type, seed_iterations, max_iterations)
    predictions = km.predict(samples)
    score = metrics.silhouette_score(samples, km.labels_, metric='euclidean')
    # Silhouette Score (-1:1): The best value is 1 and the worst value is -1
    print "silhouette_score: {0}, Clusters: {1}, Type: {2}, Seed: {3}, Iterations: {4}". \
    format(score, num_clusters, sequence_type, seed_iterations, max_iterations)
    return (km, predictions)
    
# start k-means for k=40 (label size=400, cluster_centers=40)
# Note that we are using the reduced set from PCA as input (X_r_mod), but still using k-means with n_clusters=40
km, predictions = start_kmeans(X_r_mod, 40, 'random', 1, 1)
km, predictions = start_kmeans(X_r_mod, 40, 'random', 10, 300)
km, predictions = start_kmeans(X_r_mod, 40, 'k-means++', 1, 1)
km, predictions = start_kmeans(X_r_mod, 40, 'k-means++', 10, 300)
# http://scikit-learn.org/stable/modules/clustering.html#clustering-evaluation
# If the ground truth labels are not known, evaluation must be performed using the model itself.
```

```python
# plot the clusters for n_clusters=5 with centroids
fig, (ax1, ax2) = plt.subplots( 1,2, figsize=(14,6), sharex=True, sharey=True)

# predict based on input reduced set from PCA (X_r_mod)
# so we should have 40 centroids to plot
km, predictions = start_kmeans(X_r_mod, 40, 'k-means++', 10, 300)
yhat = km.predict(X_r_mod)
centroids = km.cluster_centers_

# plot k-means predicted and target values of clusters of the first two principal components
for t in range(0, 40):
    # given target
    ax1.scatter(X_r_mod[y == t,0],
                X_r_mod[y == t,1],
                c=cmap.colors[t])
    # predicted
    ax2.scatter(X_r_mod[yhat == t,0],
                X_r_mod[yhat == t,1],
                c=cmap.colors[t])

# titles and centroids, centroids marked with 'x'
ax1.set_title('True Value')
ax1.scatter(centroids[:,0],centroids[:,1],marker = 'x', c='b')
ax2.set_title('Predicted Value')
ax2.scatter(centroids[:,0],centroids[:,1],marker = 'x', c='b')

plt.show()
```

#### Clustering Operation Status
* No, clustering did not quite work, with a 'random' placement nor with 'k-means++' initialization type.  While increasing the number of iterations and/or seed helps, it still does not lead to the clustering function as operational.  However we can say that since we did not get any negative 'silhouette_scores', at least datapoints were not obviously assigned to wrong cluster.

#### Centroid map to single Individual
* Since the value of the silhouette_score is somewhat close to '0', this denotes overlapping clusters, so each Individual would not their own cluster (for cluster_size=40).  As there are some centroids that are mapped quite close together, and some in a cluster of the same color that were not predicted correct in comparing to the truth target value.


## Question 5. Vary the number 'k' as a function of the silhouette coefficient

```python
coef_map = []
for k in range(2, n_samples):
    km = kmeans_alg(X_r_mod, k, 'k-means++', 10, 300)
    predictions = km.predict(X_r_mod)
    c =  metrics.silhouette_score(X_r_mod, km.labels_, metric='euclidean')    
    coef_map.append([k, c])
    #print "Iteration: {0}, Coefficient: {1}".format(k, c)
```

```python
# Plot Silhouette coefficient as a function of k
fig = plt.figure(figsize=(8,5))
#plt.plot(coef_np_map[:,0], coef_np_map[:,1], 'ro-', linewidth=2)
plt.scatter(coef_np_map[:,0], coef_np_map[:,1])
plt.title('k vs. Silhouette Coefficient')
plt.xlabel('k (Number of Clusters)')
plt.ylabel('Silhouette Coefficient')
plt.xlim([0,400]) 
plt.ylim([0,0.35])
```

* Increasing the number of clusters did not have the affect in terms of an optimal coefficient for clustering, even with several iterations and placement judgement values.  A coefficient value of 1.0 would have been optimal, but instead we never reached over ~0.33.  Note this was acting on an already reduced data set from PCA.
* As k was scaled, inbetween each reached a peak at k=118, and as k was further scaled, the coefficient started to drop off noting that it was no longer performing optimal.
