```python
# Decision Trees, Random Forests
# Dataset: https://archive.ics.uci.edu/ml/datasets/Bank+Marketing
# Goal: The classification goal - predict if the client will subscribe (yes/no) to a term deposit (variable y)    
```

```python
import pandas as pd
import numpy as np

from sklearn.cross_validation import ShuffleSplit
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import tree
from IPython.core.display import Image

from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing

df = pd.read_csv("./bank/bank.csv", sep=";")
df_f = pd.read_csv("./bank-additional/bank-additional-full.csv", sep=";")
df.head()
```

### Basic Banking Set 

```python
def describe_dataset(df):
    print "Features: {0}\n".format(df.drop("y", axis=1, inplace=False).columns)
    print "Target Datatype: {0}\n".format(df.y.dtype)
    print "Feature Datatypes:\n{0}".format(df.drop("y", axis=1, inplace=False).dtypes)
    
describe_dataset(df)
```

The dataset seems to consist a mix of binary, numeric integers or floats and string categorical data.  It is implied that some data can be continuous; e.g. as a Date Field (month, day, duration), while other numeric values are clearly discrete, e.g. balance.

```python
# Transform categorical string features to numerical features
def transform_features(df):
    le = preprocessing.LabelEncoder()
    # Transform features of object type
    df_ = df.loc[:, df.dtypes == object].copy()
    labels_i = df_.columns
    print "Features to Transform:{0}".format(labels_i)
    for label in df_: df_[label] = le.fit_transform(df_[label].values)
    for label in labels_i: df[label] = df_[label]
    return df

df = transform_features(df)        
df.head()
```

```python
# Split the Dataset so as not to overtrain (70/30)   
df_features = df.drop("y", axis=1, inplace=False)
df_target = df.y    
X_train, X_test, y_train, y_test = train_test_split(df_features.values, df_target.values, test_size=0.3, random_state=0)

# Predictions: will client subscribe to a term deposit
model = tree.DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
results = model.predict(X_test)
pd.crosstab(y_test, results, rownames=['actual'], colnames=['predicted'])
```

```python
# Determine number of Mispredictions
df_features_ = pd.DataFrame(X_test, columns=df_features.columns)
df_target_   = pd.DataFrame(y_test, columns=['y (actual)'])
df_results   = pd.DataFrame(results, columns=['y (predicted)'])

# join the features, actual, and predicted into one dataframe
df_ = df_features_.join(df_target_).join(df_results) 
df_errors = df_[(df_['y (actual)'] != df_['y (predicted)'])]
print "Number of Mispredicted Records: {0}".format(len(df_errors))
```

```python
# Cross Validation with K-Folds
def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold

def cross_validate_optimal_folds(X, y, classifier, max_folds=50):
    res = []
    for k in xrange(2,max_folds):
        res.append( (k, cross_validate(X, y, classifier, k) ) )
    
    df = pd.DataFrame(res, columns=["folds", "score"])    
    print "Maximum Fold Accuracy: {0}".format(df["score"].max())
    print "\nTop 5 Filtered Sorted Folds:\n{0}".format(df.sort(['score'], ascending=[False])[:5])

# Cross Validation based on Training set
# Default evaluation with n_folds=10
model_cv =  tree.DecisionTreeClassifier(random_state=0).fit 
print "n_folds=10 CV: {0}".format(cross_validate(X_train, y_train,model_cv,10))
# Optimal n_folds for highest accuracy, while taking into account complexity
cross_validate_optimal_folds(X_train, y_train, model_cv, 25)
```

The maximum accuracy is achieved with n_folds = 7, with a score of ~0.88.  The complexity is still low, so we will choose this for n_folds.

### Ensemble Method: Random Forests

### Learning Curves
* Plot of Learning Curve for RandomForestClassifier (w/CV n_folds=15)

```python
# Training Score, Cross Validation Score as varied over number of samples
from sklearn.learning_curve import learning_curve

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and traning learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : integer, cross-validation generator, optional
        If an integer is passed, it is the number of folds (defaults to 3).
        Specific cross-validation objects can be passed, see
        sklearn.cross_validation module for the list of possible objects

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt


title = "Learning Curves (Random Forest)"
estimator = RandomForestClassifier(random_state=0)
cv = ShuffleSplit(df_features_f.shape[0], n_iter=100, test_size=0.2, random_state=0)
plot_learning_curve(estimator, title, df_features_f.values, df_target_f.values, cv=cv, n_jobs=4)
```

Notice that both for the Training and CV Curves for Random Forest, they do not have a very steep increase in score with experience (number of samples) is increased.  Both the Training and CV curves benefit from more data (experience), but not by much relative to their own curve.  

The training score is much larger than the CV score for the maximum number of samples, so adding more training samples should help increase generalization.
